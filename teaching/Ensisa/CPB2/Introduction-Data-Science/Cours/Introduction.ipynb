## **List Manipulation in Python**

- Retrieving an element from a list by indexing

```python
my_list = [1, 'house', 2.3, 'Marc', 3, 5, 'dad']

print(my_list[0]) # get first element counting forward
print(my_list[1]) # get second element counting forward

print(my_list[-1]) # get first element counting backwards
print(my_list[-2]) # get second element counting backwards
```

- Retrieving a set of elements from a list using `:` (slicing)

```python
my_list = [1, 'house', 2.3, 'Marc', 3, 5, 'dad']

print(my_list[1:3]) # get the second and third element, exclude the forth
```

- Use slicing by only setting the beginning or end

```python
my_list = [1, 'house', 2.3, 'Marc', 3, 5, 'dad']

print(my_list[:3]) # get all elements from the start until index 2, exclude index 3
print(my_list[3:]) # get all elements from index 3 (included) till the end of the list
```

```python
# donwload the football data
!wget -nc https://germain-forestier.info/dataset/foot.py

# heights and positions are available as lists
from foot import heights, positions

# Convert positions and heights to numpy arrays: np_positions, np_heights
np_heights = np.array(heights)
np_positions = np.array(positions)

# Heights of the goalkeepers: gk_heights
gk_heights = np_heights[np_positions == 'GK']

# Heights of the other players: other_heights
other_heights = np_heights[np_positions != 'GK']

# Print out the median height of goalkeepers. Replace 'None'
print("Median height of goalkeepers: " + str(np.median(gk_heights)))

# Print out the median height of other players. Replace 'None'
print("Median height of other players: " + str(np.median(other_heights)))
```
