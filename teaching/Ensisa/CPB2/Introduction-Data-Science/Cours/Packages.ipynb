 ## **Classification models**

 In Machine Learning, a classification model's task is to learn how to classify samples into different categories.

 In order to do that we are in need of

 ### **Train/Test splits**

 The goal of having a training set different than a test/evaluation test is to be able to correcly evaluate the performance of a classification model on new data (unseen)

 Example in real life: students and their exams

- When dividing a dataset into train/test sets, we should be carefull about the class distribution, we should avoid the case where almost no samples of a given class exist in the test set
- For this reason we define what is called "stratification"

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv('http://germain-forestier.info/dataset/iris.csv', header=0)
mask = np.random.rand(len(df)) < 0.8
df_train = df[mask]
df_test = df[~mask]

def plotdata(df, title):
  subset = df[df['species'] == 'setosa']
  plt.scatter(subset['sepal_length'], subset['sepal_width'], label='setosa')

  subset = df[df['species'] == 'versicolor']
  plt.scatter(subset['sepal_length'], subset['sepal_width'], label='versicolor')

  subset = df[df['species'] == 'virginica']
  plt.scatter(subset['sepal_length'], subset['sepal_width'], label='virginica')

  plt.xlabel('sepal_length')
  plt.ylabel('sepal_width')
  plt.title(title)
  plt.legend()
  plt.show()

plotdata(df_train,'Iris train')
plotdata(df_test, 'Iris test')
```

### **Confusion Matrix**

The confusion matrix allows to visualize and analyze the model's predictions, it allows to:

- Compare the predicted labels and the ground truth
- Evaluates the performance of a classification mdoel
- Find the errors confusions between classes (in the point of view of a classification model)

```python
from sklearn.metrics import confusion_matrix, classification_report, precision_score

real = ['setosa','setosa','versicolor','versicolor','virginica','virginica']
pred = ['setosa','setosa','versicolor','virginica','virginica','virginica']

print(confusion_matrix(real, pred))
print(classification_report(real , pred))
print(precision_score(real , pred, average='micro'))
```

 ### Solution - Exercise 2

```python
import pandas as pd
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt


# reading the data
df = pd.read_csv('http://germain-forestier.info/dataset/iris.csv', header=0)

# create your threshold classifier here

df['pred'] = 'setosa'
df.loc[df['petal_width'] > 0.8,'pred'] = 'versicolor'
df.loc[df['petal_width'] > 1.7,'pred'] = 'virginica'

print(confusion_matrix(df['species'], df['pred']))
print(classification_report(df['species'], df['pred']))

# plot the borders of the classes

groups = df.groupby('species')

fig, ax = plt.subplots()
ax.margins(0.05)
for name, group in groups:
    ax.plot(group.petal_length, group.petal_width, marker='o', linestyle='', ms=4, label=name)
plt.xlabel("petal_length")
plt.ylabel("petal_width")
plt.title("Iris dataset")
plt.axhline(0.8)
plt.axhline(1.7)
ax.legend()
plt.show()
```
